```python
def json_extract(obj, key):
    """Recursively fetch values from nested JSON."""
    arr = []

    def extract(obj, arr, key):
        """Recursively search for values of key in JSON tree."""
        if isinstance(obj, dict):
            for k, v in obj.items():
                if isinstance(v, (dict, list)):
                    extract(v, arr, key)
                elif k == key:
                    arr.append(v)
        elif isinstance(obj, list):
            for item in obj:
                extract(item, arr, key)
        return arr

    values = extract(obj, arr, key)
    return values

def json_chunk_extract(obj, key):
    """Recursively fetch chunks from nested JSON."""
    arr = []

    def extract(obj, arr, key):
        """Recursively search for keys in JSON tree."""
        if isinstance(obj, dict):
            if key in obj.keys():
                arr.append(obj)
            else:
                for k, v in obj.items():
                    extract(v, arr, key)
        elif isinstance(obj, list):
            for item in obj:
                extract(item, arr, key)
        return arr

    values = extract(obj, arr, key)
    return values

def flatten_json(obj):
    """Flatten chunks of json"""
    output = []
    
    def flatten(obj, output, curr_key):
        """Flatten chunks of json"""
        if isinstance(obj, dict):
            for k, v in obj.items():
                if not isinstance(v, (dict,list)):
                    output.append(f'{curr_key},{k}|{v}')
                else:
                    flatten(v, output, f'{curr_key},{k}')
        elif isinstance(obj, list):
            for item in obj:
                if not isinstance(item, (dict,list)):
                    output.append(f'{curr_key}|{item}')
                else:
                    flatten(item, output, curr_key)     
        return output

    flat_json = flatten(obj, output, '')
    return flat_json
```
